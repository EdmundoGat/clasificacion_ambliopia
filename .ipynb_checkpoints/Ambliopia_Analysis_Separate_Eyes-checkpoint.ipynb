```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("BASE_AMBLIPIA.csv") 
```

```python
cl1=data[data["DIAGNOSTICO"]=="AMBLIOPIA REFRACTIVA (ANISOMETROPICA)"]
cl2=data[data["DIAGNOSTICO"]=="AMBLIOPIA REFRACTIVA"]
frames = [cl1,cl2]
data_final = pd.concat(frames)
```

```python
catB=data_final.values[:,4]
cat=[0]*len(catB)
for i in range(625):
    if catB[i]=='AMBLIOPIA REFRACTIVA (ANISOMETROPICA)':
        cat[i]=1
```

```python
X = data_final.values[:,[8,9,10]]
scaler = StandardScaler()
scaled_X_OI = scaler.fit_transform(X)
```

```python
pca = PCA(n_components = 1)
X_OI=principalComponents = pca.fit_transform(scaled_X_OI)
```

```python
pca.explained_variance_ratio_
```

```python
X = data_final.values[:,[5,6,7]]
scaler = StandardScaler()
scaled_X_OD = scaler.fit_transform(X)
```

```python
pca = PCA(n_components = 1)
X_OD=principalComponents = pca.fit_transform(scaled_X_OD)
pca.explained_variance_ratio_
```

```python
colors = ['red','green']
fig = plt.figure()
ax = fig.gca()
plt.grid()
plt.scatter(X_OI[:,0], X_OD[:,0],c=cat, cmap=matplotlib.colors.ListedColormap(colors))

plt.show()
```

```python
X_log = np.concatenate((X_OI, X_OD), axis=1)
X_log
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_log,cat,test_size=0.25,random_state=40)


model = LogisticRegression(solver='lbfgs',max_iter=2000,multi_class='auto')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
parameters = model.coef_
accuracy
```
